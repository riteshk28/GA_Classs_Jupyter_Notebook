```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
import glob
```

```python
#important note: column names should be exact match. It make it easy to merge two datasets

p1 = 'D:\MyData\GreyAtom\Python\Jupyter\pract_data\data\data\movielens\movies.dat'
c1 = ['movie_id', 'movie', 'genre']
movies = pd.read_table(p1, delimiter='::', header=None, names=c1)
#movies.head(10)
p2 = 'D:\MyData\GreyAtom\Python\Jupyter\pract_data\data\data\movielens\users.dat'
c2 = ['user_id', 'gender', 'age_group', 'occupation', 'zip']
users = pd.read_table(p2, delimiter='::', header=None, names=c2)
#users.head(10)

p3 = 'D:\\MyData\\GreyAtom\\Python\\Jupyter\\pract_data\\data\\data\\movielens\\ratings.dat'
c3 = ['user_id', 'movie_id', 'rating', 'timestamps']
ratings = pd.read_table(p3, delimiter='::', header=None, names=c3)
#ratings.head(10)

```

```python
#we merge ratings and users first since it has user_id in common aand then movies since the data would have movie_id in common
data = pd.merge(pd.merge(ratings, users), movies)
data.head(10)
```

```python
genderized_mean = data.pivot_table('rating', aggfunc=np.mean, index='movie', columns='gender').fillna(0)
#dropna or fillna(0) may change the results
```

```python
genderized_mean.head(10)
```

```python
female_domination = genderized_mean.sort_values(by='F', ascending=False)
female_domination
```

```python
#To  get list of movies who have rating more than 2.5
good_movies = data[data['rating'] >= 2.5]
good_movies
```

```python
#to get films who has more than 250 ratings
title_group = data.groupby('movie')
most_reviewed_data = title_group['rating'].count()
most_reviewed = most_reviewed_data[most_reviewed_data >= 250].index
```

```python
most_reviewed
```

```python
popular_means = genderized_mean.loc[most_reviewed]
print popular_means.head(10)
female_popular = popular_means.sort_values(by='F', ascending=False)
print female_popular.head(10)
```

```python
popular_means['diff'] = popular_means['M'] - popular_means['F']
disagreement = popular_means.sort_values(by='diff', ascending=False).head(10)
disagreement.head()
```
